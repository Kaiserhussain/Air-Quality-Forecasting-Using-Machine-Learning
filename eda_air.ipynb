```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = r"C:\Users\Bhargavi\Desktop\Python\city_data.csv"
data = pd.read_csv(file_path)
```

```python
# Initial Data Inspection
print("Initial Data Inspection:")
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
# Check for Missing Values
print("Checking for Missing Values:")
print(data.isnull().sum())
```

```python
# Data Visualization

# Histogram
print("Plotting Histograms...")
data.hist(bins=30, figsize=(15, 10))
plt.show()
```

```python
# Scatter Plot
# Replace 'PM2.5' and 'AQI' with the correct feature names
if 'PM2.5' in data.columns and 'AQI' in data.columns:
    print("Plotting Scatter Plot...")
    sns.scatterplot(x='PM2.5', y='AQI', data=data)
    plt.title('Scatter Plot of PM2.5 vs AQI')
    plt.show()
else:
    print("PM2.5 or AQI not found in the dataset.")
```

```python
# Heatmap
print("Plotting Heatmap...")

# Select only numeric columns for correlation matrix
numeric_data = data.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Identify and Handle Outliers
# Replace 'PM2.5' with the correct feature name
if 'PM2.5' in data.columns:
    print("Plotting Boxplot...")
    sns.boxplot(data['PM2.5'])
    plt.title('Boxplot for PM2.5')
    plt.show()
else:
    print("PM2.5 not found in the dataset.")
```
